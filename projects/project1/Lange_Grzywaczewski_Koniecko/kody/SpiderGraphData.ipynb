# Data for spider graph

```python
import fastf1
import numpy as np
import pandas as pd

from IPython.display import clear_output
from time import sleep
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
from matplotlib.colors import ListedColormap
from matplotlib import cm
```

```python
fastf1.Cache.enable_cache("./cache")
```

```python
# Na ten moment jest 19 zakoÅ„czonych sesji GP
SESSIONS = [i for i in range(1, 20)]
```

```python
# Kryteria
# high_wind: 
```

```python
race_name = []
wind_max = []
air_max = []
pressure_max = []
rain_perc = []
redbull_points = []
mercedes_points = []
ferrari_points = []
```

```python
def get_points_by_team(res, team_name) -> int:
    return res[res.TeamName == team_name].Points.sum()
```

```python
for ses_id in SESSIONS:
    ses = fastf1.get_session(2022, ses_id, "R")
    ses.load(messages = False)
    clear_output()
    race_name.append(ses.event.name)
    wd = ses.weather_data
    wind_max.append(wd.WindSpeed.max())
    air_max.append(wd.AirTemp.max())
    pressure_max.append(wd.Pressure.max())
    rain_perc.append(wd.Rainfall.sum() / len(wd.Rainfall))
    res = ses.results
    mercedes_points.append(get_points_by_team(res, "Mercedes"))
    ferrari_points.append(get_points_by_team(res, "Ferrari"))
    redbull_points.append(get_points_by_team(res, "Red Bull Racing"))
```

```python
SESSIONS_RES = {
    "WindSpeedMax": wind_max,
    "AirTempMax": air_max,
    "PressureMax": pressure_max,
    "Rain": rain_perc,
    "Redbull": redbull_points,
    "Mercedes": mercedes_points,
    "Ferrari": ferrari_points
}
```

```python
df = pd.DataFrame(SESSIONS_RES, index = race_name)
```

```python
df.Rain = df.Rain.round(2)
```
